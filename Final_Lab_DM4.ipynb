```python
# Part 1 â€” Basic Function Arguments with User Input

# Function definition
def describe_pet(animal_type, pet_name):
    print(f"I have a {animal_type} and its name is {pet_name}.")

# Ask the user to input details for three pets
for i in range(3):
    print(f"\nEnter details for pet #{i+1}:")
    animal = input("Enter animal type: ")
    name = input("Enter pet name: ")
    describe_pet(animal, name)
```

```python
#Part 2 Positional vs Keyword Arguments

# Positional arguments
describe_pet("rabbit", "Thumper")

# Keyword arguments (order reversed)
describe_pet(pet_name="Whiskers", animal_type="hamster")
```

```python
#Part 3  Default Arguments

# Function with default argument
def describe_pet(pet_name, animal_type="dog"):
    print(f"I have a {animal_type} and its name is {pet_name}.")

# Calls using default and custom values
describe_pet("Brownie")            # animal_type defaults to dog
describe_pet("Nugget", "chicken")  # specifying both arguments
```

```python
#Part 4 Create Your Own Function order_drink

# Function definition
def order_drink(drink, size="medium", iced=False):
    # Decide text based on iced parameter
    ice_text = "with ice" if iced else "no ice"
    return f"Your order: {size} {drink} {ice_text}"

# 1. Order a default drink (medium coffee, no ice)
print(order_drink("coffee"))

# 2. Order a large hot chocolate (no ice)
print(order_drink("chocolate", size="large"))

# 3. Order a small iced milk tea (iced=True)
print(order_drink("milk tea", size="small", iced=True))
```

```python
#Part 5 Mini Challenge Calculator

# Function definition
def compute(operation, num1, num2=1):
    if operation == "add":
        return num1 + num2
    elif operation == "multiply":
        return num1 * num2
    elif operation == "subtract":
        return num1 - num2
    else:
        return "Invalid operation"

# Tasks
print(compute("add", 5, 10))                 
print(compute("multiply", num1=3, num2=4))  
print(compute("subtract", 20))                #19  (20 - default 1)
print(compute("divide", 10, 2))              # Invalid 
```
